```python
import numpy as np
import pandas as pd 
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv('ak91.csv')
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns = ['quarter_of_birth'], dtype=int)
```

```python
df.head()
```

```python
df.rename(columns={'quarter_of_birth_1.0': 'Q1', 'quarter_of_birth_2.0': 'Q2', 'quarter_of_birth_3.0': 'Q3', 'quarter_of_birth_4.0': 'Q4'}, inplace=True)
```

```python
df.head()
```

```python
model = smf.ols(formula= 'years_of_schooling~Q4 + C(state_of_birth) + C(year_of_birth 	)', data = df).fit()
```

```python
print(model.summary())
```

```python
df['edu_return'] = model.fittedvalues
```
