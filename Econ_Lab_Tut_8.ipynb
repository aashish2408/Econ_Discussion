```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm
```

```python
x=np.sort(np.random.uniform(0,4*np.pi,200))
```

```python
e = np.random.uniform(0,1,200)
```

```python
y = np.cos(x) + e
```

```python
plt.scatter(x,y)
plt.show()
```

2

```python

from statsmodels.nonparametric.kernel_regression import KernelReg
```

```python
ksrmv = KernelReg(y,x, var_type='c',bw=[0.1] )
pred_y, marginal_effects = ksrmv.fit(x)
```

```python
plt.scatter(x,y)
plt.plot(x,pred_y)
```

```python
ksrmv = KernelReg(y,x, var_type='c',bw=[0.5] )
pred_y, marginal_effects = ksrmv.fit(x)
plt.scatter(x,y)
plt.plot(x,pred_y)
```
