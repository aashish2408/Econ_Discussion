```python
t = 0.5
n = 100
B = 10000
z = []
f_hat_t = np.mean(x<=t)
```

```python
bootstrap_samples = np.random.choice(x, (B, n), replace=True)
f_hat_star = np.mean(bootstrap_samples <= t, axis=1)
```

```python
# Calculate the bootstrap statistic
bootstrap_stat = np.sqrt(n) * (f_hat_star - f_hat_t)

# Compare with Normal distribution
sigma = np.sqrt(f_hat_t * (1 - f_hat_t))
```

```python
plt.hist(bootstrap_stat, bins = 30, density = True)
```

```python
plt.figure(figsize = (8,6))
plt.hist(bootstrap_stat, bins=30, density=True, alpha=0.6, color='blue', label='Bootstrap Statistic')
```

```python
from scipy.stats import norm
plt.figure(figsize = (8,6))
plt.hist(bootstrap_stat, bins=30, density=True, alpha=0.6, color='blue', label='Bootstrap Statistic')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
plt.plot(x, norm.pdf(x, 0, sigma), 'r--', label='N(0, )')
plt.xlabel('Statistic Value')
plt.ylabel('Density')
plt.title('Comparison of Bootstrap Statistic with Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```
