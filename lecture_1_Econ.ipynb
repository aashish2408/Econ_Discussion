```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

X= np.random.uniform(1,4,(100,10000))
M = np.mean(np.mean(X,axis=0))-2.5
V = np.mean(np.var(X,axis=0))-.75

print('The bias for mean is: {}'.format(M))
print('The bias for variance is: {}'.format(V))
```

```python
dff1=np.zeros(20)
dff2=np.zeros(20)
eps=.02

for i in range(20):
    X= np.random.uniform(1,4,(500*(i+1),10000))
    H1 = (np.sign(np.abs(np.mean(X,axis=0)-2.5)-eps)+1)/2 
    H2 = (np.sign(np.abs(np.var(X,axis=0)-.75)-eps)+1)/2
    dff1[i]= np.mean(H1)  
    dff2[i]= np.mean(H2)
plt.plot(np.array(range(500, 10500,500)),dff1,'b')
plt.plot(np.array(range(500, 10500,500)),dff2,'g')

plt.show()
```

```python
x_axis = np.arange(-3, 3, 0.01)

n=200
X=np.random.uniform(1,4,(n,10000))
clt1 = np.sqrt(n)*(np.mean(X,axis=0)-2.5)
clt2 = np.sqrt(n)*(np.var(X,axis=0)-.75*.99)

fig, axs = plt.subplots(1, 2, figsize=(13, 5))

axs[0].hist(clt1, bins=50, density= True)
axs[0].plot(x_axis, norm.pdf(x_axis, 0, np.sqrt(.75)))
axs[0].set_title("CLT for mean")

axs[1].hist(clt2, bins=50, density= True)
axs[1].plot(x_axis, norm.pdf(x_axis, 0, np.sqrt(.45)))
axs[1].set_title("CLT for variance")
```
